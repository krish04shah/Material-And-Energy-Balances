```python
import numpy as np
from plotly.subplots import make_subplots
```

```python
R = 8.314 #J/mol*K
Tc = 647.096 #K
Pc = 22.064e6 #Pa
```

```python
a = (27/64)*(((R**2)*(Tc**2))/(Pc))
b = (1/8)*((R*Tc)/Pc)
```

```python
def vdwP(V,T):
  e = 0.
  s = 0.
  return (R*T/(V-b)) - (a/((V-e*b)*(V-s*b)))
```

```python
V = np.logspace(np.log10(1.01*b),-2, 1000)
```

```python
fig1 = make_subplots()

for T in np.arange(275,901,25):
  fig1.add_scatter(x = V, y = vdwP(V,T),mode = 'lines',name = f'{T} K')

fig1.update_xaxes(type = 'log')
fig1.update_yaxes(range = [-1e8,1e8])
fig1.update_layout(width = 800, height = 600)

fig1
```
